# Plot Graphs

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/LavanyaPolamarasetty/Datasets/master/Classification/Social_Network_Ads.csv")
df
```

```python
#Line Plot
plt.title("Line Plot")
plt.plot(df['EstimatedSalary'])
plt.show()
```

```python
#Scatter plot
plt.title("Scatter Plot")
plt.scatter(df["Age"],df["EstimatedSalary"])
plt.xlabel("Age")
plt.ylabel("Estimated Salary")
plt.show()
```

```python
#Histogram
plt.title("Histogram")
plt.hist(df["EstimatedSalary"])
plt.xlabel("EstimatedSalary")
plt.show()
```

```python
#Bar graph
plt.title("Bar Graph")
plt.bar(df["Purchased"],df["Age"])
plt.xlabel("Purchased")
plt.ylabel("Age")
plt.show()
```

```python
#Pie Chart 
plt.title("Pie Chart")
plt.pie(df["Age"][:10],labels = df["Gender"][:10], autopct = "% 1.1f%%")
plt.show()
```

```python
#Box Plot
plt.title("Box Plot")
plt.boxplot(df["User ID"])
plt.show()
```

```python
#Subplots
fig, area = plt.subplots(2, 2)
area[0][0].plot(df['Gender'])
area[0][1].plot(df['Purchased'])
area[1][0].plot(df['EstimatedSalary'])
area[1][1].plot(df['Age'])
plt.show()
```

```python
plt.subplot(1, 2, 2)
plt.plot(df['Age'])
plt.subplot(1, 2, 1)
plt.plot(df['Gender'])
plt.show()
```
